```python
import urllib.request
import pandas as pd
import boto3
from io import BytesIO
import io
import pyarrow.parquet as pq
from tabulate import tabulate
```

```python
def extract_data (url, filename):
  try:
    urllib.request.urlretrieve(url, filename)
  except Exception as e:
    print(e)
```

```python
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/c9509ab4-6f6d-4b97-979a-0cf2a10c922b/download/tmphrybkxuh.csv", "data/data_2015.csv")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/b7ea6b1b-3ca4-4c5b-9713-6dc1db52379a/download/tmpzxzxeqfb.csv", "data/data_2016.csv")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/30022137-709d-465e-baae-ca155b51927d/download/tmpzccn8u4q.csv", "data/data_2017.csv")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/2be28d90-3a90-4af1-a3f6-f28c1e25880a/download/tmp7602cia8.csv", "data/data_2018.csv")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/ea2e4696-4a2d-429c-9807-d02eb92e0222/download/tmpcje3ep_w.csv", "data/data_2019.csv")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/6ff6a6fd-3141-4440-a880-6f60a37fe789/download/tmpcv_10m2s.csv", "data/data_2020.csv")
```

```python
arquivos = [
    "data/data_2015.csv",
    "data/data_2016.csv",
    "data/data_2017.csv",
    "data/data_2018.csv",
    "data/data_2019.csv",
    "data/data_2020.csv"
]
```

```python
df = {}
```

```python
# buscando o ano de cada arquivo e armazenando no df

for arquivo in arquivos:
  ano = arquivo.split("_")[-1].split(".")[0]
  df[ano] = pd.read_csv(arquivo)
```

Conectando ao AWS S3

```python
aws_access_key_id = ""
aws_secret_access_key = ""
region_name = ""

boto3.setup_default_session(
    aws_access_key_id = aws_access_key_id,
    aws_secret_access_key = aws_secret_access_key,
    region_name = region_name
)

s3 = boto3.client("s3")
```

Convertendo CSV para PARQUET e importando dados no S3

```python
for ano, dfs in df.items():
  parquet_buffer = BytesIO()
  dfs.to_parquet(parquet_buffer)

  s3.put_object(
      Bucket = "alr-s3-datalake-boston",
      Key = f"bronze/dados_{ano}.parquet",
      Body = parquet_buffer.getvalue()
  )
```

Validando

```python
response = s3.list_objects(Bucket = "alr-s3-datalake-boston")
```

```python
keys = [obj["Key"] for obj in response ["Contents"]]
print(keys)
```

Acessando dados

```python
bucket_name = 'alr-s3-datalake-boston'
prefix = 'bronze/'

# Listando arquivos na pasta bronze
response = s3.list_objects_v2(Bucket=bucket_name, Prefix=prefix)

# Inicializando uma lista para armazenar DataFrames
df_bronze = []

# Verificando se há arquivos na resposta
if 'Contents' in response:
    print("Arquivos na camada bronze:")

    # Iterando sobre os arquivos listados
    for obj in response['Contents']:
        file_key = obj['Key']

        # Garantindo que estamos processando apenas arquivos Parquet
        if file_key.endswith('.parquet'):
            print(f"Processando arquivo: {file_key}")

            # Fazendo download do arquivo Parquet
            response = s3.get_object(Bucket=bucket_name, Key=file_key)

            # Lendo o arquivo Parquet diretamente do buffer em um DataFrame
            file_stream = io.BytesIO(response['Body'].read())
            df1 = pq.read_table(file_stream).to_pandas()

            # Adicionando o DataFrame à lista
            df_bronze.append(df1)

    # Concatenando todos os DataFrames em um único DataFrame
    if df_bronze:
        final_df = pd.concat(df_bronze, ignore_index=True)
        print("Dados combinados com sucesso!")
        print(tabulate(final_df.head(), headers='keys', tablefmt='psql'))
    else:
        print("Nenhum arquivo Parquet encontrado.")
else:
    print("Nenhum arquivo encontrado na camada bronze.")
```
